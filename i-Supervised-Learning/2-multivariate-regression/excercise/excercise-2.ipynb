```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from word2number import w2n

pd.set_option('display.float_format', lambda x: '%.3f' % x)
pd.options.display.float_format = '{:,.2f}'.format
```

```python
df = pd.read_csv('hiring.csv')
df
```

```python
df.experience = df.experience.fillna('zero')
df
```

```python
import math
median_score = math.floor(df['test_score(out of 10)'].median())
```

```python
df['test_score(out of 10)'].fillna(median_score, inplace=True)
df
```

```python
df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience',	'test_score(out of 10)', 'interview_score(out of 10)']], df['salary($)'])
reg.predict([[10, 10, 10]])
# How much would you pay me if I had 10 years of experience, full test score, and full interview score?
```
