##### The above is a coditional probability where it is basically shown that we need to find the probability of survival based on certain conditions such as Gende, Class, Age, Cabin, Fare. This is a case of conditonal probability.

##### We call this a naive bayes, because we make a naive assumption that these features, as in gender, class, age etc are independent of each other.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 

COLOR = 'white'
plt.rcParams['text.color'] = COLOR
plt.rcParams['axes.labelcolor'] = COLOR
plt.rcParams['xtick.color'] = COLOR
plt.rcParams['ytick.color'] = COLOR

pd.set_option('display.float_format', lambda x: '%.3f' % x)
pd.options.display.float_format = '{:,.2f}'.format
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df.drop(['PassengerId', 'Name', 'Parch', 'Cabin', 'Ticket', 'Embarked'], axis='columns', inplace=True)
df

target = df.Survived
inputs = df.drop('Survived', axis='columns')
```

```python
dummies = pd.get_dummies(inputs.Sex)
dummies.head()
```

```python
inputs = pd.concat([inputs, dummies], axis='columns')
```

```python
inputs.drop('Sex', inplace=True, axis='columns')
inputs
```

```python
inputs
```

```python
inputs
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
```

```python
inputs.isna().sum()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
