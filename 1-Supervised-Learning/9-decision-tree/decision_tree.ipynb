```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

COLOR = 'white'
plt.rcParams['text.color'] = COLOR
plt.rcParams['axes.labelcolor'] = COLOR
plt.rcParams['xtick.color'] = COLOR
plt.rcParams['ytick.color'] = COLOR

pd.set_option('display.float_format', lambda x: '%.3f' % x)
pd.options.display.float_format = '{:,.2f}'.format
```

```python
df = pd.read_csv('salaries.csv')
df.head()
```

```python
#  target variable are going to be taken as inputs
inputs = df.drop('salary_more_then_100k',axis = 'columns')
inputs
```

```python
target = df['salary_more_then_100k']
target
```

```python
#  Machine learning algos cant understand words, only numbers, so turn them into numbers 
# We have many columns as words so h\we have to create multiple LabelEncoder() objects for each one.
```

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_le'] = le_company.fit_transform(inputs['company'])
inputs
```

```python
inputs['degree_le'] = le_company.fit_transform(inputs['degree'])
inputs['job_le'] = le_company.fit_transform(inputs['job'])
inputs
```

```python
inputs_le = inputs.drop(['company', 'degree', 'job'],axis='columns')
```

```python
inputs_le
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()
```

```python
X = inputs_le
y = target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state= 10)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
